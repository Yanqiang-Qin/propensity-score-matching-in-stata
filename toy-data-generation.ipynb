# TOY DATA

```r
source("http://data.vanderbilt.edu/tgs/misc/r_functions.txt")
require(Hmisc)
require(knitr)
```

```r

set.seed(2340)
N <- 1e5
data <- data.frame(
    id               = 1:N,
    age              = sample(18:75,          N, replace = TRUE, prob = 1/(5 + abs(18:75 - 55))),
    sex              = sample(c("M","F"),     N, replace = TRUE),
    race             = sample(c("W","B","O"), N, replace = TRUE, prob = c(3,2,1)),
    diabetes         = sample(c("Y","N"),     N, replace = TRUE, prob = c(2,4)),
    disease_severity = rchisq(N,2)
)
```

## CATEGORICAL VARIABLES

```r
table1 <- summaryM(age + sex + race + diabetes + disease_severity ~ 1, data = data)
print(table1, exclude1 = FALSE, long = TRUE, digits = 2)
```

## AGE VARIABLE

```r
mypar2()
par(mar = c(2,2,2,1))

# PLOT AGE DISTRIBUTION
plot(
    ecdf(data$age), 
    main = "", 
    xlab = "AGE"
)

age_quantiles <- quantile(data$age, 1:3/4)

abline(
    v = age_quantiles, 
    lwd = 3, 
    col = "red"
)

axis(
    3, 
    at = age_quantiles, 
    labels = "Q" %|% 1:3
)

hist(
    data$age, 
    freq = FALSE, 
    xlab = "AGE", 
    main = "",
    col  = "lightblue"
)
box()
```

## DISEASE SEVERITY SCORE

```r
mypar2()
par(mar = c(2,2,2,1))

# PLOT AGE DISTRIBUTION
plot(
    ecdf(data$disease_severity), 
    main = "", 
    xlab = "DISEASE SEVERITY SCORE",
    xlim = range(data$disease_severity)
)

disease_severity_quantiles <- quantile(data$disease_severity, 1:3/4)

abline(
    v = disease_severity_quantiles, 
    lwd = 3, 
    col = "red"
)

axis(
    3, 
    at = disease_severity_quantiles, 
    labels = "Q" %|% 1:3
)

hist(
    data$disease_severity, 
    freq = FALSE, 
    xlab = "DISEASE SEVERITY", 
    main = "",
    col  = "lightblue"
)
box()
```

## ASSIGN TREATMENT

Treatment will be assigned as a function of disease severity.  Patients with higher disease scores will be more likely to receive treatment 1 (instead of treatment 0).

The following plot indicates the probability of assignment by disease severity.

```r
sigmoid <- function(x) 1/(1 + exp(-x))
probability_treatment <- sigmoid(log(1+data$disease_severity) - mean(log(1+data$disease_severity)))
data$treatment <- rbinom(N, 1, probability_treatment)
mypar2()
plot(
    data$disease_severity, 
    probability_treatment,
    xlab = "DISEASE SEVERITY",
    ylab = "PROBABILITY OF TREATMENT"
)
```

## OUTCOME SCORE

The outcome score will be a function of age, disease severity, and treatment.

$$
E[Y] = 9 + 1 * \textrm{log}(\textrm{age}) + 2 * \textrm{disease_severity} - 10 * \textrm{treatment 1}
$$


```r
mean_response <- 9 + log(data$age) + 2 * data$disease_severity - 10 * data$treatment
error_term <- rnorm(N, 0, 2)
data$outcome <- mean_response + error_term
```

## NAIVE COMPARISON OF OUTCOME AND TREATMENT

```r
mypar2()
plot_ecdf <- function(x){
    plot(
        ecdf(x$outcome), 
        add = TRUE, 
        col = x$treatment + 1, 
        lwd = 3
    )
    points(median(x$outcome), .5, col = "white", pch = 16, cex = 4)
    points(median(x$outcome), .5, col = x$treatment + 1, pch = 1, cex = 4)
    text(median(x$outcome), .5, x$treatment, cex = 2, col = x$treatment + 1)
}

plot.new()
plot.window(xlim = range(data$outcome), ylim = c(0,1))
garbage <- by(data, data$treatment, plot_ecdf)
box()
axis(1)
axis(2)
lm1 <- lm(outcome ~ treatment, data = data)
summary(lm1)
```
